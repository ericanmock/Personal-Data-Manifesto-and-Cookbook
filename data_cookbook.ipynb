## Recipe #1: How to Merge Two Dataframes

Merging dataframes can be extremely helpful!! Often times merging, bringing together relevant information from different sources into a single, unified dataset, will allow for greater data analysis. 
On the other hand, sometimes data is so big that the best way of displaying it is through subsetting the data and separating it. 

In this example, we will merge the match results df from 2009 with the match results df from 2023. That way, we can see all of the matches and access all of the data from these two specific years. 



The function .merge() combines data from two DataFrames based on a common column or set of columns (called the "merge key"). It can also perform various types of joins like inner join, left join, right join, and outer join depending on how you want to handle missing data.

For more information on .merge() visit https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.merge.html .

```python
# use .merge with key = 'tourney name'; I only want to see the results
# from tournaments that took place in both years.

stats = stats_2009.merge(stats_2023, on = 'tourney_name')
stats.head()
```

As we can see here, the original column names have changed by including and _x or _y  at the end of the column names. That is because these names are repreated in both datasets, so this is pandas way of letting us which name belongs to which set. All except tourney_name which is just tourney_name because it is the merge key, so there is only one column titled tourney_name.

The following code simply turns the columns into a list and displays it. I find this helpful when I want to look at all of the datapoints we have access to, but not the actual values. 


```python
# displaying columns
column_names = stats.columns.tolist()
print(column_names)
```

## Recipe #2: How to Filter DFs: two different ways


Filtering dataframes is essential for data organization! By filtering, we can isolate specific data points relevant to our analysis, ignoring irrelevant noise. That way, we can uncover hidden patterns, improve accuracy, simplify analysis, and save processing time. 


We can also perform calculations, manipulate data, and build models, if we wanted to.



In this example, I will filter my big stats df two different ways to include only final round results (as in final rounds of the tournament because every match played is part of a tournement). I can do this simply by creating a new df.

```python
# first create df that only contains final round results
# there are 2 ways to do this, first: 

final_rounds = stats[(stats['round_x'] == 'F') & (stats['round_y'] == 'F')]

# second:

#final_rounds = stats[stats['round_x'].isin(['F']) & stats['round_y'].isin(['F'])]

final_rounds.head()
```

## Recipe #3: Use a for loop to access values & use .value_counts()

In this next chunk of code, I will be creating a new df that only contains the tournament ids and winner names. I am getting even MORE selective. I only care about the players who are actually capable of winning a tournament. 

Then, I use a for loop to iterate through the df and print out the tournament id and corresponding winner name. 

```python
# modify df to only contain columns 'tourney_id_x', 'tourney_id_y',
# 'winner_name_x', 'winner_name_y'

tourney_winners = final_rounds[['tourney_id_x', 'tourney_id_y','winner_name_x', 'winner_name_y']]

# access the winners 
for index, row in tourney_winners.iterrows():
    
    #tournaments
    tourney_id_2009 = row['tourney_id_x']
    tourney_id_2023 = row['tourney_id_y']
    
    #winners
    winner_09 = row['winner_name_x']
    winner_23 = row['winner_name_y']
    
    #print(f"2009 Tournament: {tourney_id_2009}, Winner: {winner_09}, 2023 Tournament:{tourney_id_2023} Winner: {winner_23}")
    
    
```

Next, I use the .value_counts() function from pandas. This function counts the number of times each unique value appears in a pandas DataFrame column. It returns a neatly sorted Series showing the frequency of each value. 

For more information about .value_counts() visit https://pandas.pydata.org/docs/reference/api/pandas.Series.value_counts.html

```python
# Count final rounds won by each player

winner_counts_09 = tourney_winners['winner_name_x'].value_counts()
winner_counts_23 = tourney_winners['winner_name_y'].value_counts()

# Filter the players who have won at least 3 final rounds

multiple_wins_09 = winner_counts_09[winner_counts_09 >= 3]
multiple_wins_23 = winner_counts_23[winner_counts_23 >= 3]

```

As you can see, I used value counts and then filtered the df for players who have won at least three tournaments. I did this because the players who have won more than two tournamnets in a single year are likely among the best in the world at the time. 

## Recipe #4: How to Make a Horizontal Bar Chart using .barh()

Time for the most fun part! Visualizations! 


Data visualizations captivate our brains and guide decision-making with clarity. They are an amazing tool to turn information into knowledge which is an essential part of the data science process.

For this example, I chose to make a horizontal bar chart to dispay the number of tournaments the top players have won. Bar charts are nice to look at and relevant for many different datasets. 

Form more information on charts in matplotlib check out: https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.barh.html

Let's get into it!

We begin this code by creating a figure (the overall plot) with two subplots (individual graphs) arranged side-by-side. The figure size is set to 10x6 inches for better viewing.

We are using multiple_wins_09 and multiple_wins_23 from the previous code. These variables are assumed to be Pandas Series objects that contain the player names as indexes and their win counts (>= 3) as values for each year.

First, we make a plot for the 09 stats by creating a horizontal bar graph on ax1. The index of multiple_wins_09 will become the y-axis labels (players), and the values will become the bar heights (win counts). The color is set to blue and a legend label is added.

Then, we create the plot for the 23 stats by mirroring the process for ax1, but with the data for 2023, green color, and a different legend label.

Finally, we display the plot.

```python
# Create figure and subplots
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 6))

# Plot bar graph for 2009 data
ax1.barh(multiple_wins_09.index, multiple_wins_09.values, color='blue', label='Wins (2009)')
ax1.set_xlabel('Tournament Wins')
ax1.set_ylabel('Player')
ax1.set_title('2009 - Players with Wins >= 3')
ax1.legend()

# Plot bar graph for 2023 data
ax2.barh(multiple_wins_23.index, multiple_wins_23.values, color='green', label='Wins (2023)')
ax2.set_xlabel('Tournament Wins')
ax2.set_ylabel('Player')
ax2.set_title('2023 - Players with Wins >= 3')
ax2.legend()

plt.tight_layout()
plt.show()
```

These plots are showing us that in 2009 there were more professional players that won more than three tournaments in 2009 than in 2023. In 2009, it looks like there were 5 players with more than three tournamnet wins, but in 2023, there were only three players with more than three tournament wins.
